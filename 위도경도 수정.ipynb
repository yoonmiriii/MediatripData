```python
import pandas as pd
```

```python
df = pd.read_csv('./media_area.csv')
```

```python
 def split_address(address):
    parts = address.split(' ')
    if len(parts) > 1:
        return parts[0], parts[1]
    else:
        return address, ''
```

```python
df[['주소_1', '주소_2']] = df['주소'].apply(split_address).tolist()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
filtered_df = df[df['주소_1'] == '경기도']
```

```python
filtered_df = filtered_df[filtered_df['미디어유형'] == '드라마']
```

```python
filtered_df = filtered_df[filtered_df['장소유형'] == '다채로운 지역 탐방']
```

```python
filtered_df = df['위도']
```

```python
filtered_df = df['경도']
```

```python
# 고유한 위도와 경도 값들을 얻기
unique_latitudes = filtered_df['위도'].unique()
unique_longitudes = filtered_df['경도'].unique()

# 조건에 맞는 행을 선택
latitude_condition = unique_latitudes >= 39
longitude_condition = unique_longitudes >= 128

filtered_df_latitude = filtered_df.loc[filtered_df['위도'].isin(unique_latitudes[latitude_condition])]
filtered_df_longitude = filtered_df.loc[filtered_df['경도'].isin(unique_longitudes[longitude_condition])]
```
